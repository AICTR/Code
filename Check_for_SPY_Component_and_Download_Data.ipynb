```python
import numpy as np
import pandas as pd
from tools import DownloadData
import json
import yfinance as yf
from datetime import datetime

import warnings
warnings.filterwarnings('ignore')
```

```python
with open("sp500-historical-components.json",'r') as load_f:
    load_dict = json.load(load_f)
```

```python
d = {}
for i in load_dict:
    if i['Date'] == '2017/05/03':
        continue
    j = i['Date'].replace('/','-')
    d[j] = i['Symbols']
    
```

```python
tmp = list(d.keys())
tmp.sort()
tmp
```

```python
dd = set()
for i,v in d.items():
    dd = dd.union(set(v))
# dd = list(dd)
```

```python
cc = dd
for i,v in d.items():
    if ((i<'2012-01-01') | (i>'2016-12-31')):
        continue
    cc = cc.intersection(v)
cc = list(cc)
```

```python
len(cc)
```

```python
s = cc[0]
s
```

```python
f = None
for s in cc:
    data = pd.read_csv('./data1/'+s+'.csv', index_col=0)[['Close']]
    data = data/data.shift(1)
    data.dropna(inplace = True)
    data.head()

    if f is None:
        f = data.copy()
        f.columns = [s]
    else:
        f[s] = data.Close
f.head()
```

```python
f.dropna(axis=1, inplace=True)
```

```python
f.to_csv('spy500_12_16.csv')
```
